# 1. Definição do Problema

O tabagismo é reconhecido como um dos principais fatores de risco para diversas doenças crônicas, estando associado a problemas cardiovasculares, respiratórios e diferentes tipos de câncer. Embora os impactos do hábito de fumar sejam amplamente conhecidos, muitas das alterações causadas pelo tabagismo podem ser observadas por meio de características físicas e exames clínicos rotineiros.

O dataset escolhido, Body Signal of Smoking, contém informações anônimas de indivíduos, incluindo características corporais, sinais vitais e resultados de exames laboratoriais. Além disso, o conjunto de dados possui uma variável que identifica se cada indivíduo é fumante ou não fumante.

O objetivo deste trabalho é investigar a relação entre indicadores de saúde e o hábito de fumar, além de desenvolver um modelo capaz de prever o status de fumante de um indivíduo a partir dessas informações. Para isso, serão realizadas etapas de análise exploratória, preparação dos dados, treinamento e avaliação de modelos de Machine Learning.


##1.1 Hipóteses do Problema

As hipóteses consideradas para este estudo são:

*   Existem diferenças mensuráveis entre fumantes e não fumantes em variáveis clínicas e corporais.
*   Alguns exames laboratoriais apresentam maior capacidade de distinguir fumantes de não fumantes.

*   É possível desenvolver um modelo de classificação com desempenho satisfatório utilizando apenas os atributos disponíveis no dataset.

*   Nem todas as variáveis possuem a mesma relevância para a identificação do hábito de fumar.

##1.2 Tipo de Problema

Este é um problema de aprendizado **supervisionado** de **classificação** binária. A variável-alvo possui duas categorias possíveis, indicando se o indivíduo é fumante ou não fumante. A partir dos atributos físicos e clínicos disponíveis, os modelos serão treinados para aprender padrões capazes de realizar essa classificação em novos indivíduos.

##1.3 Seleção dos Dados

##2.1 Atributos do Dataset
Os atributos escolhidos representam diferentes aspectos da saúde dos indivíduos, incluindo características demográficas, composição corporal, indicadores cardiovasculares, exames laboratoriais e condições odontológicas. Essa diversidade de informações pode contribuir para a identificação de padrões associados ao hábito de fumar e para o treinamento de modelos preditivos mais robustos.


* **ID**: Identificador único de cada registro no conjunto de dados.

* **age**: Idade do indivíduo em anos.

* **gender**: Sexo biológico do indivíduo [M = masculino; F = feminino].

* **height(cm)**: Altura do indivíduo em centímetros.

* **weight(kg)**: Peso corporal em quilogramas.

* **waist(cm)**: Circunferência da cintura em centímetros, utilizada como indicador de gordura abdominal.

* **eyesight(left)**: Acuidade visual do olho esquerdo.

* **eyesight(right)**: Acuidade visual do olho direito.

* **hearing(left)**: Capacidade auditiva do ouvido esquerdo [1 = normal; 2 = alteração auditiva].

* **hearing(right)**: Capacidade auditiva do ouvido direito [1 = normal; 2 = alteração auditiva].

* **systolic**: Pressão arterial sistólica, correspondente à pressão exercida nas artérias durante a contração do coração.

* **relaxation**: Pressão arterial diastólica, correspondente à pressão exercida nas artérias durante o relaxamento do coração.

* **fasting blood sugar**: Nível de glicose no sangue após período de jejum, utilizado como indicador de alterações metabólicas.

* **Cholesterol**: Quantidade total de colesterol presente no sangue.

* **triglyceride**: Concentração de triglicerídeos no sangue, relacionada ao metabolismo de gorduras.

* **HDL**: Lipoproteína de alta densidade, conhecida como "colesterol bom".

* **LDL**: Lipoproteína de baixa densidade, conhecida como "colesterol ruim".

* **hemoglobin**: Concentração de hemoglobina no sangue, responsável pelo transporte de oxigênio no organismo.

* **Urine protein**: Quantidade de proteína detectada na urina, utilizada como indicador de possíveis alterações renais.

* **serum creatinine**: Nível de creatinina sérica, utilizado para avaliação da função renal.

* **AST**: Aspartato aminotransferase, enzima utilizada na avaliação da função hepática.

* **ALT**: Alanina aminotransferase, enzima associada à saúde do fígado.

* **Gtp**: Gama-glutamil transferase, enzima frequentemente relacionada ao consumo de álcool e alterações hepáticas.

* **oral**: Status do exame oral realizado no indivíduo.

* **dental caries**: Presença de cáries dentárias [0 = não; 1 = sim].

* **tartar**: Presença de tártaro dentário [Y = sim; N = não].

* **smoking**: Status de fumante do indivíduo [0 = não fumante; 1 = fumante]. **Esta é a variável-alvo do problema de classificação.**

##4.1 Separação entre Variáveis Explicativas e Variável-Alvo

```python
X = df.drop('smoking', axis=1)
y = df['smoking']

print("Dimensão de X:", X.shape)
print("Dimensão de y:", y.shape)
```

Nesta etapa, o conjunto de dados foi dividido entre as variáveis explicativas (X), utilizadas para prever o hábito de fumar, e a variável-alvo (y), representada pela coluna smoking.

##4.2 Divisão em Treino e Teste


```python
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42,
    stratify=y
)

print("Treino:", X_train.shape)
print("Teste:", X_test.shape)
```

Para avaliar a capacidade de generalização dos modelos, o conjunto de dados foi dividido em conjuntos de treino e teste utilizando a estratégia holdout. Foram destinados 80% dos registros para treinamento dos algoritmos e 20% para teste.

Após a divisão, o conjunto de treino passou a conter 44.553 observações, enquanto o conjunto de teste ficou composto por 11.139 observações. Essa separação permite que os modelos sejam treinados em uma parcela dos dados e avaliados em registros não utilizados durante o treinamento, fornecendo uma estimativa mais realista do desempenho em novos dados.

A utilização do parâmetro stratify=y garantiu a preservação da proporção original de fumantes e não fumantes nos conjuntos de treino e teste, evitando distorções na distribuição da variável-alvo durante o processo de divisão dos dados.

##4.3 Padronização dos Dados

Alguns algoritmos de Machine Learning são sensíveis à escala das variáveis, especialmente aqueles baseados em distância ou otimização matemática, como Regressão Logística, KNN e SVM. Dessa forma, foi realizada a padronização dos atributos utilizando o método StandardScaler.

A padronização transforma as variáveis para uma escala com média igual a zero e desvio padrão igual a um, evitando que atributos com valores naturalmente maiores exerçam influência desproporcional sobre o modelo.

O ajuste do escalonador foi realizado apenas com os dados de treinamento, sendo posteriormente aplicado aos dados de teste, evitando vazamento de informação (data leakage).

```python
# Criando o escalonador
scaler = StandardScaler()

# Ajustando apenas nos dados de treino
X_train_scaled = scaler.fit_transform(X_train)

# Aplicando a transformação nos dados de teste
X_test_scaled = scaler.transform(X_test)
```

```python
pd.DataFrame(
    X_train_scaled,
    columns=X_train.columns
).agg(['mean', 'std'])
```

A análise das estatísticas descritivas após a aplicação do StandardScaler confirmou a correta padronização dos dados. Observa-se que as variáveis passaram a apresentar médias próximas de zero e desvios padrão próximos de um, indicando que a transformação foi realizada com sucesso.

Embora algumas variáveis continuem apresentando valores extremos, esses registros foram mantidos por possivelmente representarem condições clínicas reais dos indivíduos avaliados. Além disso, algoritmos baseados em árvores de decisão tendem a ser menos sensíveis à presença de outliers.

##4.4 Estratégia de Validação

Além da divisão dos dados em conjuntos de treino e teste, foi adotada uma estratégia de validação cruzada para obter uma estimativa mais robusta do desempenho dos modelos.

A técnica escolhida foi a validação cruzada estratificada (Stratified K-Fold), que divide os dados em diferentes subconjuntos preservando a proporção original das classes da variável-alvo. Essa abordagem reduz a dependência de uma única divisão treino-teste e permite avaliar a estabilidade dos modelos em diferentes amostras dos dados.

```python
# Definindo a estratégia de validação cruzada

cv = StratifiedKFold(
    n_splits=5,
    shuffle=True,
    random_state=42
)

cv
```

Foi utilizada a configuração de cinco partições (5-fold), amplamente empregada em problemas de classificação. Dessa forma, cada modelo será treinado e avaliado em diferentes subconjuntos dos dados, permitindo uma estimativa mais confiável do seu desempenho e capacidade de generalização.

O objetivo do baseline é fornecer uma referência mínima de desempenho para comparação com os demais modelos. Neste trabalho foi utilizada a estratégia most_frequent, que sempre prevê a classe majoritária do conjunto de treinamento.

##5.2 Definição dos Modelos

Para este trabalho foram selecionados cinco algoritmos de classificação com características distintas:

- Regressão Logística: modelo linear amplamente utilizado em problemas de classificação binária.
- K-Nearest Neighbors (KNN): algoritmo baseado na proximidade entre observações.
- Random Forest: método ensemble baseado em múltiplas árvores de decisão.
- Support Vector Machine (SVM): algoritmo que busca encontrar a melhor fronteira de separação entre as classes.
- XGBoost: técnica de gradient boosting amplamente utilizada em competições e aplicações de Machine Learning devido ao seu elevado desempenho.

Os algoritmos foram selecionados por representarem diferentes abordagens de aprendizado supervisionado, permitindo comparar modelos lineares, baseados em distância, ensembles e métodos de boosting.

```python
# Criação dos modelos

models = {
    'Logistic Regression': LogisticRegression(
        max_iter=1000,
        random_state=42
    ),

    'KNN': KNeighborsClassifier(),

    'Random Forest': RandomForestClassifier(
        random_state=42
    ),

    'SVM': SVC(
        random_state=42
    ),

    'XGBoost': XGBClassifier(
        random_state=42,
        eval_metric='logloss'
    )
}
```

##5.3 Validação Cruzada dos Modelos

Para comparar os modelos de forma consistente, foi utilizada validação cruzada estratificada com cinco partições (5-fold). Como métrica inicial de comparação foi utilizada a acurácia, que representa a proporção de classificações corretas realizadas pelo modelo.

```python
# Avaliação dos modelos por validação cruzada

results = []

for name, model in models.items():

    scores = cross_val_score(
        model,
        X_train_scaled,
        y_train,
        cv=cv,
        scoring='accuracy'
    )

    results.append([
        name,
        scores.mean(),
        scores.std()
    ])

# Criando tabela de resultados

results_df = pd.DataFrame(
    results,
    columns=[
        'Modelo',
        'Acurácia Média',
        'Desvio Padrão'
    ]
)

results_df.sort_values(
    by='Acurácia Média',
    ascending=False
)
```

Os resultados da validação cruzada indicam que o modelo Random Forest apresentou o melhor desempenho entre os algoritmos avaliados, alcançando uma acurácia média de 81,7% e o menor desvio padrão entre os folds, evidenciando elevada estabilidade e capacidade de generalização.

O modelo XGBoost apresentou o segundo melhor desempenho, com acurácia média de 78,1%, seguido pelo SVM (76,4%), Regressão Logística (74,5%) e KNN (73,0%).

Os resultados sugerem que modelos baseados em árvores de decisão foram mais eficientes na identificação do hábito de fumar do que modelos lineares ou baseados em distância. Isso indica que as relações entre as variáveis clínicas e a variável-alvo apresentam características não lineares, que são melhor capturadas por algoritmos ensemble como Random Forest e XGBoost.

Considerando os resultados obtidos, os modelos Random Forest e XGBoost serão analisados de forma mais detalhada nas próximas etapas por meio de métricas adicionais de classificação.

##6.4 Análise da Capacidade de Generalização

Após a avaliação das métricas no conjunto de teste, foi realizada uma análise da capacidade de generalização dos modelos. Essa etapa consiste em comparar o desempenho obtido nos conjuntos de treinamento e teste, permitindo identificar possíveis sinais de sobreajuste (overfitting).

O overfitting ocorre quando o modelo aprende excessivamente os padrões específicos dos dados de treinamento, apresentando desempenho muito elevado durante o treinamento, mas reduzindo sua capacidade de realizar previsões em novos dados.

Para essa análise, foram calculadas as acurácias dos modelos Random Forest e XGBoost no conjunto de treinamento e comparadas com os resultados obtidos no conjunto de teste.


```python
# Previsões no conjunto de treinamento

y_pred_train_rf = rf.predict(X_train_scaled)
y_pred_train_xgb = xgb.predict(X_train_scaled)

# Acurácia de treinamento

train_acc_rf = accuracy_score(y_train, y_pred_train_rf)
train_acc_xgb = accuracy_score(y_train, y_pred_train_xgb)

# Tabela comparativa

generalizacao = pd.DataFrame({

    'Modelo': ['Random Forest', 'XGBoost'],

    'Accuracy Treino': [
        train_acc_rf,
        train_acc_xgb
    ],

    'Accuracy Teste': [
        accuracy_score(y_test, y_pred_rf),
        accuracy_score(y_test, y_pred_xgb)
    ]
})

generalizacao
```

##6.5 Ajuste de Hiperparâmetros do Random Forest

Após a análise da capacidade de generalização, observou-se que o modelo Random Forest apresentou acurácia de 100% no conjunto de treinamento e 82,88% no conjunto de teste, indicando sinais de sobreajuste (overfitting).

Com o objetivo de reduzir esse comportamento e buscar uma melhor capacidade de generalização, foi realizado um ajuste de hiperparâmetros utilizando a técnica Grid Search associada à validação cruzada estratificada.

Foram avaliadas diferentes combinações de parâmetros relacionados à complexidade do modelo, incluindo o número de árvores (n_estimators), profundidade máxima das árvores (max_depth), número mínimo de amostras para divisão de nós (min_samples_split) e número mínimo de amostras nas folhas (min_samples_leaf).

```python
param_grid = {

    'n_estimators': [100, 200],

    'max_depth': [10, 20, None],

    'min_samples_split': [2, 5],

    'min_samples_leaf': [1, 2]

}

grid_rf = GridSearchCV(

    estimator=RandomForestClassifier(random_state=42),

    param_grid=param_grid,

    cv=5,

    scoring='accuracy',

    n_jobs=-1

)

grid_rf.fit(X_train_scaled, y_train)

best_rf = grid_rf.best_estimator_
```

```python
# Melhor modelo encontrado pelo Grid Search

rf_otimizado = best_rf

# Treinando o modelo otimizado
rf_otimizado.fit(X_train_scaled, y_train)

# Previsões no conjunto de teste
y_pred_rf = rf_otimizado.predict(X_test_scaled)

# Probabilidades para ROC-AUC
y_prob_rf = rf_otimizado.predict_proba(X_test_scaled)[:, 1]

# Accuracy do modelo otimizado
acc_otimizado = accuracy_score(
    y_test,
    y_pred_rf
)

print("Melhores parâmetros:")
print(grid_rf.best_params_)

print("\nMelhor acurácia CV:")
print(grid_rf.best_score_)

print("\nAccuracy otimizado:")
print(acc_otimizado)
```

A aplicação do Grid Search permitiu identificar a combinação de hiperparâmetros que apresentou o melhor desempenho durante a validação cruzada. Os melhores parâmetros encontrados foram profundidade máxima das árvores igual a 20 (`max_depth=20`), número mínimo de amostras por folha igual a 1 (`min_samples_leaf=1`), número mínimo de amostras para divisão dos nós igual a 2 (`min_samples_split=2`) e 200 árvores na floresta (`n_estimators=200`).

A melhor acurácia média obtida durante a validação cruzada foi de 82,03%. Após o treinamento do modelo otimizado e sua avaliação no conjunto de teste, foi obtida acurácia de 82,94%.

Comparando com o modelo Random Forest original, que apresentou acurácia de 82,88%, observa-se um ganho discreto de desempenho após o ajuste dos hiperparâmetros. Embora a melhoria tenha sido pequena, os resultados indicam que a configuração otimizada apresentou desempenho superior, sendo selecionada como modelo final para as análises subsequentes.


##6.6 Matriz de Confusão do Random Forest

A matriz de confusão permite analisar detalhadamente os acertos e erros do modelo de classificação. Por meio dela é possível identificar quantos indivíduos foram corretamente classificados como fumantes e não fumantes, bem como os erros cometidos em cada classe.

Nesta etapa foi utilizada a versão otimizada do modelo Random Forest, obtida após o ajuste de hiperparâmetros. As linhas representam as classes reais e as colunas representam as classes previstas pelo modelo.

##6.7 Curva ROC e ROC-AUC

A Curva ROC (Receiver Operating Characteristic) é uma ferramenta gráfica utilizada para avaliar a capacidade de discriminação dos modelos de classificação. Ela representa a relação entre a taxa de verdadeiros positivos (Recall) e a taxa de falsos positivos para diferentes limiares de decisão.

Além da curva, é calculada a métrica ROC-AUC (Area Under the Curve), que quantifica a capacidade do modelo em distinguir corretamente as classes. Quanto mais próximo de 1 for o valor da AUC, melhor é o desempenho do classificador.

```python
# Probabilidades da classe positiva (fumante)

y_prob_rf = rf_otimizado.predict_proba(X_test_scaled)[:, 1]

y_prob_xgb = xgb.predict_proba(X_test_scaled)[:, 1]

# Curva ROC - Random Forest

fpr_rf, tpr_rf, _ = roc_curve(
    y_test,
    y_prob_rf
)

auc_rf = auc(
    fpr_rf,
    tpr_rf
)

# Curva ROC - XGBoost

fpr_xgb, tpr_xgb, _ = roc_curve(
    y_test,
    y_prob_xgb
)

auc_xgb = auc(
    fpr_xgb,
    tpr_xgb
)

# Gráfico

plt.figure(figsize=(8,6))

plt.plot(
    fpr_rf,
    tpr_rf,
    linewidth=2,
    label=f'Random Forest (AUC = {auc_rf:.3f})'
)

plt.plot(
    fpr_xgb,
    tpr_xgb,
    linewidth=2,
    label=f'XGBoost (AUC = {auc_xgb:.3f})'
)

plt.plot(
    [0,1],
    [0,1],
    linestyle='--',
    color='gray',
    label='Classificador Aleatório'
)

plt.title(
    'Gráfico 7: Curva ROC dos Modelos Avaliados',
    fontweight='bold'
)

plt.xlabel('Taxa de Falsos Positivos (FPR)')
plt.ylabel('Taxa de Verdadeiros Positivos (TPR)')

plt.legend()
plt.grid(alpha=0.3)

plt.tight_layout()

plt.show()

print(f'ROC-AUC Random Forest: {auc_rf:.4f}')
print(f'ROC-AUC XGBoost: {auc_xgb:.4f}')
```

Os resultados da Curva ROC reforçam o desempenho superior do modelo Random Forest em relação ao XGBoost.

O Random Forest apresentou ROC-AUC de 0,9129, indicando excelente capacidade de discriminação entre fumantes e não fumantes. Esse resultado significa que o modelo possui elevada probabilidade de atribuir escores maiores aos indivíduos fumantes do que aos não fumantes, demonstrando forte capacidade de separação entre as classes.

O XGBoost também apresentou desempenho satisfatório, alcançando ROC-AUC de 0,8618. Embora esse valor indique boa capacidade de classificação, seu desempenho foi inferior ao observado no Random Forest.

A superioridade do Random Forest observada na ROC-AUC é consistente com os resultados obtidos nas métricas de Accuracy, Precision, Recall e F1-Score, bem como na análise da matriz de confusão. Dessa forma, o modelo Random Forest otimizado foi confirmado como a melhor solução para o problema de predição do hábito de fumar na base de dados analisada.
